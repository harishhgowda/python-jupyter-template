### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.figure(figsize=(8, 6))
n, bins, patches = plt.hist(data, 
                           bins=5,  # Number of bins
                           color='skyblue', 
                           alpha=0.8, 
                           edgecolor='black', 
                           linewidth=1.5)

# Add title and labels
plt.title('Customized Histogram', fontsize=16, fontweight='bold')
plt.xlabel('Data Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Add frequency labels on top of each bar
for i in range(len(patches)):
    # Get the height of each bar
    height = patches[i].get_height()
    if height > 0:  # Only add label if there's a bar
        plt.text(patches[i].get_x() + patches[i].get_width()/2, height + 0.05,
                 f'{int(height)}', ha='center', va='bottom', fontweight='bold')

# Add grid for better readability
plt.grid(axis='y', alpha=0.3)

# Set y-axis to show integer values only
plt.yticks(range(0, int(max(n)) + 2))

# Display the plot
plt.tight_layout()
plt.show()
```
