### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='red', s=100, alpha=0.7, edgecolors='black', linewidth=2)

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.annotate(f'({x[i]}, {y[i]})', 
                (x[i], y[i]), 
                xytext=(10, 10),  # Offset the annotation
                textcoords='offset points',
                fontsize=10,
                ha='left',
                bbox=dict(boxstyle='round,pad=0.3', facecolor='yellow', alpha=0.7))

# Add title and labels
plt.title('Scatter Plot with Annotations', fontsize=16, fontweight='bold')
plt.xlabel('X Values', fontsize=12)
plt.ylabel('Y Values', fontsize=12)

# Add grid for better readability
plt.grid(True, alpha=0.3)

# Set axis limits for better visualization
plt.xlim(0.5, 5.5)
plt.ylim(1, 11)

# Display the plot
plt.tight_layout()
plt.show()
```
